## Importando bibliotecas

```python
import pandas as pd
```

## Lendo dataframe

```python
df = pd.read_csv('metadados_filmes.csv')
```

## Calculando quantil passado por usuário para feature vote_count

```python
d = float(input("Digite um valor para o quantil entre 0.0 e 1: "))
m = df['vote_count'].quantile(d)
m
```

## Filtragem por duração dos filmes

```python
run_time_select = df[(df['runtime'] >= 45) & (df['runtime'] <= 300)]
run_time_select
```

## Filtragem apenas por filmes que receberam mais de m votos

```python
votos_m_select = run_time_select[(run_time_select['vote_count'] >= m)]
votos_m_select
```

## Validando Shape

```python
votos_m_select.shape
```

## Calculando média de vote_average e atribuindo a C

```python
C = df['vote_average'].mean()
C
```

## Criando função de classificação ponderada

```python
def classificacao_ponderada(x, m=m, C=C):
  v = x['vote_count']
  R = x['vote_average']
  return (v/(v+m) * R) + (m/(m+v) * C)
```

## Cálculo da pontuação e adição da coluna 'score

```python
votos_m_select['score'] = votos_m_select.apply(classificacao_ponderada, axis=1)
votos_m_select['score']
```

## Ordenando dataframe de forma decrescente com base na feature score

```python
df_ordenado_decrescente = votos_m_select.sort_values(by='score', ascending=False)
df_ordenado_decrescente
```

## Listando os 25 melhores filmes

```python
top_25_filmes = df_ordenado_decrescente[['title', 'vote_count', 'vote_average', 'score', 'runtime']].head(25)
top_25_filmes
```
