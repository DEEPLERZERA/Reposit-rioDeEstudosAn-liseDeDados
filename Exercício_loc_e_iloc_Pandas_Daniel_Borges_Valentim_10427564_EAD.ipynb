## Importando biblioteca

```python
import pandas as pd
```

## Criando Dataframe

```python
df = pd.DataFrame({
    'Nome':['Alice', 'Bob', 'Charlie', 'David'],
    'Idade':[25,30,35,40]
})
```

```python
df
```

## Fazendo filtro de idade

```python
df.loc[df['Idade']>25]
```

```python
df.loc[(df['Idade']>25) & (df['Idade']<40)]
```

## Fazendo um slice da coluna

```python
resultado = df.loc[1:2, 'Nome':'Idade']
resultado
```

## Verificando tipo da coluna

```python
type(resultado)
```

## Imprimindo o nome no ndice 2

```python
df.loc[2, 'Nome']
```

```python
df.loc[2, 'Nome'] = 'Seu Shima'
df
```

## Acessando elementos individuais com iloc

```python
df.iloc[0,0]
```

```python
df.iloc[3,0]
```

```python
df.iloc[[1,3],[0,1]]
```

## Fatiamento (slicing) com .iloc

```python
df.iloc[1:3,:]
```

```python
df.iloc[1:,:]
```

## Filtragem de linhas com .iloc

```python
df.iloc[(df["Idade"]>30).values]
```
